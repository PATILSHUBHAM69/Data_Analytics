```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import metrics
sn.set()
```

```python
iris_data = pd.read_csv("Iris.csv")
iris_data.head()
```

```python
iris_data.describe()
```

```python
iris_data['Species'].value_counts()
```

```python
plt.title('Species Count')
sn.countplot(iris_data['Species'])
```

```python
# Q1 Comparison between various species based on sepal length and width using respective visualization
```

```python
plt.figure()
sn.scatterplot(iris_data['SepalLengthCm'],iris_data['SepalWidthCm'],hue = iris_data['Species'],s=100)
```

```python
# Q2 Perform the Bi-variate Analysis via pairplot visualization
```

```python
sn.pairplot(iris_data,hue="Species")
```

```python
# Q3 Justify Sepal Length and Sepal Width features are slightly correlated with each other
```

```python
plt.figure()
sn.heatmap(iris_data.corr(),annot=True)
plt.plot()
```

```python
# Q4 Visualize the categorical feature Species with all other four input variables
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20,10))
sn.boxplot( y="PetalWidthCm", x= "Species", data=iris_data, orient='v' , ax=axes[0, 0])
sn.boxplot( y="PetalLengthCm", x=  "Species", data=iris_data, orient='v' , ax=axes[0, 1])
sn.boxplot( y="SepalLengthCm", x=  "Species", data=iris_data, orient='v' , ax=axes[1, 0])
sn.boxplot( y="SepalWidthCm", x=  "Species", data=iris_data, orient='v' , ax=axes[1, 1])
plt.show()
```

```python
# Q4 Plot the density of the length and width in the species.
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20,10))
sn.violinplot( y="PetalWidthCm", x= "Species", data=iris_data, orient='v' , ax=axes[0, 0],inner='quartile')
sn.violinplot( y="PetalLengthCm", x= "Species", data=iris_data, orient='v' , ax=axes[0, 1],inner='quartile')
sn.violinplot( y="SepalLengthCm", x= "Species", data=iris_data, orient='v' , ax=axes[1, 0],inner='quartile')
sn.violinplot( y="SepalWidthCm", x= "Species", data=iris_data, orient='v' , ax=axes[1, 1],inner='quartile')
plt.show()
```

```python
# Q6 Plot the probability density function(PDF) with each feature as a variable on X-axis and its histogram and corresponding kernel density plot on Y-axis.
```

```python
sn.FacetGrid(iris_data, hue="Species", height=5) \
.map(sn.distplot, "SepalLengthCm") \
.add_legend()

sn.FacetGrid(iris_data, hue="Species", height=5) \
.map(sn.distplot, "SepalWidthCm") \
.add_legend()

sn.FacetGrid(iris_data, hue="Species", height=5) \
.map(sn.distplot, "PetalLengthCm") \
.add_legend()

sn.FacetGrid(iris_data, hue="Species", height=5) \
.map(sn.distplot, "PetalWidthCm") \
.add_legend()
plt.show()
```
